#**1.Varience Treshold**

- Varyans Eşiği, modellemede pek kullanılmayan tüm düşük varyans özelliklerini veri kümesinden kaldıran bir özellik seçicidir.
- İstenilen çıktılara (y) değil, yalnızca özelliklere (x) bakar ve bu nedenle denetimsiz öğrenme için kullanılabilir.


Eşik Varsayılan Değeri 0

- Varyans Eşiği = 0 ise (Sabit Özellikleri Kaldır)
- Varyans Eşiği > 0 ise (Yarı-Sabit Özellikleri Kaldır)

```python
#Python Uygulaması:
import pandas as pd
import numpy as np
from sklearn.feature_selection import VarianceThreshold

# Loading data from train.csv file
df= pd.read_csv("/gdrive/My Drive/DataMining/kddcup99_csv.csv")
df.head(5)
```

```python
df.describe()
```

```python
df['lnum_outbound_cmds'].value_counts()
df.drop('lnum_outbound_cmds', axis=1, inplace=True)
df['is_host_login'].value_counts()
df.drop('is_host_login', axis=1, inplace=True)
```

```python
df['protocol_type'] = df['protocol_type'].astype('category')
df['service'] = df['service'].astype('category')
df['flag'] = df['flag'].astype('category')
cat_columns = df.select_dtypes(['category']).columns
df[cat_columns] = df[cat_columns].apply(lambda x: x.cat.codes)


X=df.drop(labels=['label'], axis=1)
y=df['label']
```

```python
var_thres=VarianceThreshold(threshold=0)

var_thres.fit(X,y)
```

```python
from sklearn.model_selection import train_test_split
# separate dataset into train and test
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3,random_state=0)

X_train.shape, X_test.shape
```

```python
var_thres.get_support()
```

```python
### Finding non constant features
sum(var_thres.get_support())
```

```python
# Lets Find non-constant features 
len(X_train.columns[var_thres.get_support()])
```

```python
constant_columns = [column for column in X_train.columns
                    if column not in X_train.columns[var_thres.get_support()]]

print(len(constant_columns))
```

```python
for column in constant_columns:
    print(column)
```

```python

X_train=X_train.drop(constant_columns,axis=1)
X_train
```

```python
abc = AdaBoostClassifier(n_estimators=50,
                         learning_rate=1)

# Train Adaboost Classifer
model = abc.fit(X_train, y_train)

#Predict the response for test dataset
y_pred = model.predict(X_test)

# Model Accuracy, how often is the classifier correct?
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```
